```python
import cooler
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib.gridspec import GridSpec
import cooltools
import bioframe
import pandas as pd
```

```python
df = pd.read_csv('GM12878.csv')
```

```python
cool_path = '/media/lchumpitaz/New Volume/Data/Hi-C/mcools/GM12878_inSitu_MboI.hg19.mapq_30.1000.mcool'
```

```python
resolution = 50000
```

```python
c=  cooler.Cooler(cool_path + '::resolutions/' + str(resolution))
```

```python
chr1, chr2 = 'chr12', 'chr17'
```

```python
mat = c.matrix().fetch(chr1,chr2)
mat2 = np.outer(df[df.chrom==chr1]['E1'].values, df[df.chrom==chr2]['E1'].values)
```

```python
fig = plt.figure(1, figsize=(20, 8))
ax1= plt.subplot(121)
im1 = ax1.matshow(mat, cmap='bwr', norm=LogNorm(vmin= (1.5) * 1e-6 , vmax=1e-4))
ax1.set_ylabel(chr1)
ax1.set_xlabel(chr2)
cbar1 = ax1.figure.colorbar(im1)
cbar1.outline.set_visible(False)

cbar1.ax.set_ylabel("α", rotation=0, va="bottom")
ax2= plt.subplot(122)
im2 = ax2.matshow(mat2, cmap='bwr', vmin = -1, vmax = 1)
ax2.set_ylabel(chr1)
ax2.set_xlabel(chr2)
cbar2 = ax2.figure.colorbar(im2)
cbar2.outline.set_visible(False)
```
